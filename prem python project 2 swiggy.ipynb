# CITY HAVING A HIGHEST AND LEAST NUMBER OF RESTAURANT

```python
city= df.City.value_counts()

plt.bar(city.index,city.values, color='red')

plt.title("Restaurant counts by city",fontdict={'fontweight':'bold'})
plt.xticks(rotation='vertical')
plt.xlabel("Cities")
plt.ylabel("Counts")

for i,j in enumerate(city.values):
    plt.text(i,j,str(j),ha='center')
    
plt.show()    
```
